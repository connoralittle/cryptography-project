```python
def login(pake, **kwargs):
    try:
        test_protocol(pake, **kwargs)
        return True
    except:
        return False
```

```python
def test_protocol(pake, **kwargs):
    print(f"Testing protocol {pake}")
    pake1 = pake(**kwargs)
    pake2 = pake(**kwargs)

    key = pake1.generate_parameters()
    pake2.set_parameters(*key)
    
    initiate_pass = pake1.initiate_pass
    response_pass = pake2.response_pass
    even_rounds = len(initiate_pass) == len(response_pass)
    symmetric = pake1.symmetric

    if even_rounds:
        ans = ()
        ans2 = ()
        for i in range(len(initiate_pass)):
            if symmetric:
                ans_temp = initiate_pass[i](*ans2)
                ans2 = response_pass[i](*ans)
                ans = ans_temp
            else:
                ans = initiate_pass[i](*ans2)
                if not ans:
                    break
                ans2 = response_pass[i](*ans)
                if not ans2:
                    break

    else:
        max_len = len(initiate_pass)
        ans2 = ()
        for i in range(max_len - 1):
            if symmetric:
                ans_temp = initiate_pass[i](*ans2)
                ans2 = response_pass[i](*ans)
                ans = ans_temp
            else:
                ans = initiate_pass[i](*ans2)
                if not ans:
                    break
                ans2 = response_pass[i](*ans)
                if not ans2:
                    break
        if ans2:
            initiate_pass[-1](*ans2)

    print("Session Key Established")
    print("Login Successful")
    print(f"Session Key A: {pake1.get_session_key()}")
    print(f"Session Key B: {pake2.get_session_key()}")
```
